# `pidgy.emoji` support



```python
:smile: `pidgy` brings support for `emoji` in tangling and weaving code.
the full experience comes to :light_bulb: in `pidgy` mode. it provides
the completion support for emojis and the ability to use `emoji` variables in python code.



`emoji` can carry many meanings & can be practical ways to avoid naming challenges in programming.

> :warning: `pidgy.emoji` can be used as a standlone extension to bring `emoji` support to `IPython`, but `pidgy` gives the full experience
```

`pidgy` permits both `emoji` and their aliases, enclosed in colons, as ways to reference 
objects with `emoji`

```python
    import pandas as :panda_face:
    import pandas as 🐼
```

```python
now `panda_face` is defined as a variables; the leading and trailing colons are stripped
```

```python
`emoji` completion helps in naming, emoji are converted to their aliases, and they
can be combined to build more complex names

    :bear::beer::baseball::ballet_shoes: = ":sleeping:"

this combination of emoji creates the name `bearbeerbaseballballet_shoes`

    assert bearbeerbaseballballet_shoes == ":sleeping:"
    
other valid characters, including unicode, can be include between `emoji`

    :capital_abcd:α:nut_and_bolt:__:hammer: = ":toolbox:"
    assert :capital_abcd:α:nut_and_bolt:__:hammer: == capital_abcdαnut_and_bolt__hammer
```

## edge cases

there are few places where the use of the emoji`":*:"` pattern

```python
### annotations

the `pidgy.emoji` transformer does conflict annotations, the colons
are stripped from the variable name and the reference is the smiley alias.

    :baby_angel:: int = 0
                
the name will be munged 
                
    assert ":baby_angel:" not in __annotations__
    __annotations__["baby_angel"]
```

```python
### slices

so far there don't seem to be any issues with `emoji` and `slice` syntax
```
