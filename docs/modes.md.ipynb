# `pidgy` modes

`pidgy` is designed primarily for `jupyter` interfaces meaning that it works well in `jupyterlab` and `notebook`.
there a few modes that a `jupyter` can interact with `pidgy` through:
 
1. traditional notebooks

  from the positional of traditional notebooks, those bearing the `.ipynb` extension, `pidgy` 
  can be used in two ways:

  1. `IPython` `%%pidgy` magic
    
      an opt-in mode triggered by `%%reload_ext pidgy.magic` that will only load the `%%pidgy`, `%%parse`, and `%%tangle` cell magics
      
      
        
  2. `IPython` pidgy extension `%%reload_ext pidgy`
    
      as an extension, all cells become `pidgy` that accept markdown input that translates to python
        
2. hybrid `pidgy` documents with `pidgy` kernel

the `pidgy` kernel turns all computation in `IPython` into `pidgy` markdown that translates to python.
the kernel can used in several modes:

  1. in a `jupyter` notebook
  
  
  
  2. in a `jupyter` console
  
  
  
  3. in a `jupyter` text editor
    
      
  
    1. with a selection based console
    
      
    
    2. with live markdown previewing
